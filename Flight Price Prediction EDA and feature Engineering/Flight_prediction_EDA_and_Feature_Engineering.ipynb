Flight Price Prediction (EDA and feature engineering)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df=train_df._append(test_df)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
## Feature Engineering Process
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head(2)
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

## Spliting Arrival time

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].apply(lambda x:x.split(":")[0])
```

```python
final_df['Arrival_Minutes']=final_df['Arrival_Time'].apply(lambda x:x.split(":")[1])
```

```python
final_df.head()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minutes']=final_df['Arrival_Minutes'].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
final_df['Dep_Time']=final_df['Dep_Time'].apply(lambda x:x.split(" ")[0])
```

```python
final_df.head()
```

```python
final_df['Dep_Hour']=final_df['Dep_Time'].apply(lambda x:x.split(":")[0])
final_df['Dep_Minutes']=final_df['Dep_Time'].apply(lambda x:x.split(":")[1])
```

```python
final_df['Dep_Hour']=final_df['Dep_Hour'].astype(int)
final_df['Dep_Minutes']=final_df['Dep_Minutes'].astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['Duration_hour']=final_df['Duration'].apply(lambda x:x.split(" ")[0].split("h")[0])
```

```python
final_df['Duration_hour']=final_df['Duration_hour'].astype('int')
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df['Airline'].unique()
```

```python
from sklearn import preprocessing 
  
# label_encoder object knows  
# how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
final_df['Airline']= label_encoder.fit_transform(final_df['Airline'])
final_df['Source']= label_encoder.fit_transform(final_df['Source']) 
final_df['Destination']= label_encoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']= label_encoder.fit_transform(final_df['Additional_Info'])
  
final_df['Airline'].unique()
final_df['Source'].unique()
final_df['Destination'].unique()
final_df['Additional_Info'].unique()
```
